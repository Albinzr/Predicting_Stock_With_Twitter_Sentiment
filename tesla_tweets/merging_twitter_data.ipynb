```python
# Import necessary libraries

import pandas as pd
```

```python
# Function to create date_time column and turn into datetime object

def fix_table_dates(dataframe):
    dataframe['date_time'] = dataframe.date + ' ' + dataframe.time
    dataframe.date_time = pd.to_datetime(dataframe.date_time)
    dataframe = dataframe.sort_values(by=['date_time'])
    dataframe = dataframe.drop(columns = ['date', 'time'])
    dataframe = dataframe.reset_index(drop=True)
    return dataframe
```

## January 2018

```python
# Import and preview 'january_2018_tweets_1.csv' (Tweets from 01/01/2018-01/14/2018)

january_2018_tweets_1 = pd.read_csv('january_2018_tweets_1.csv')
january_2018_tweets_1.head()
```

```python
january_2018_tweets_1.groupby(['date'])['tweet'].count()
```

```python
# Clean date data for january_2018_tweets_1

january_2018_tweets_1 = fix_table_dates(january_2018_tweets_1)
```

```python
# Import and preview 'january_2018_tweets_2.csv' (Tweets from 01/15/2018-01/31/2018)

january_2018_tweets_2 = pd.read_csv('january_2018_tweets_2.csv')
january_2018_tweets_2.head()
```

```python
# Clean date data for january_2018_tweets_2

january_2018_tweets_2 = fix_table_dates(january_2018_tweets_2)
```

```python
# Merge January 2018 Tesla tweets (january_2018_tweets_1 and january_2018_tweets_2) into one dataframe

january_2018_tweets = pd.concat([january_2018_tweets_1, january_2018_tweets_2], ignore_index=True)
```
